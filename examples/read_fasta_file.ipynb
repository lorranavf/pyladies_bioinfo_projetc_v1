```python
from entities.file import FileHandler
```

```python
def test_file_handler(path: str):
    sequences = FileHandler.read(path)

    print('Testing FileHandler.read() method')
    print(f'Total sequences extracted: {len(sequences)}')
    print()

    for seq in sequences:
        print(f'ID: {seq.id}')
        print(f'Description: {seq.description}')
        print(f'Sequence: {seq.sequence}')
        print(f'Alphabet: {seq.sequence.which_alphabet()}')
        print(f'Length: {seq.sequence.length()}')
        print('---')
```

```python
print('Reading Protein sequences file:')
test_file_handler('data/ptn_sequences.fasta')
    
```

```python
print('Reading RNA sequences file:')
test_file_handler('data/rna_sequences.fasta')
```

```python
print('Reading DNA sequences file:')
test_file_handler('data/dna_sequences.fasta')
```
